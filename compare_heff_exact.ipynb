```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from math import*
from numpy import linalg as LA
#L = 4
#SEED = 2022
X,Y,Z = np.loadtxt('new_eigenvalues_data.txt', dtype="complex",delimiter = '\t',unpack=True)
delta = X.real
effective = Y.real
exact = Z.real
```

```python
import os
os.environ["PATH"] += os.pathsep + '/Library/TeX/texbin'
L = 4
SEED = 4000

Page_value = str(np.around((L/2)*np.log(2)-0.5,3))

with plt.style.context(['science', 'scatter','grid']):
    fig, ax = plt.subplots(figsize=(12, 10))
    #plt.plot(delta,exact,'b--')
    plt.scatter(delta, exact, s=20,marker = 'o',c = 'blue',label=r'$\phi_{\rm exact}$')
    plt.scatter(delta, effective, s=30, marker = '+',c = 'red',label=r'$E_{\rm eff}$')

    #lgd = r"L$=12$"
    #ax.legend(title=lgd, loc=4, ncol=1)
    ax.set_title(r"L$="+str(L)+"$, Seed = $"+str(SEED)+"$, Page Value $"+str(Page_value)
                 +"$",size=26, pad=10)
    xlbl = r"Noise $\delta$"
    ylbl = r"Energy $\phi_{F}$"
    ax.set_xlabel(xlbl,size = 26)
    ax.set_ylabel(ylbl,size = 26)
    ax.set_xlim([0, 0.3])
    ax.set_ylim([-3.14, 3.14])
    #ax = f.gca()
    ax.tick_params(axis="x", direction="inout", length=8, width=1, color="k")
    ax.tick_params(axis="y", direction="inout", length=8, width=1, color="k")
    plt.xticks(fontsize=30, rotation=0)
    plt.yticks(fontsize=30, rotation=0)
    plt.legend(loc='upper left',fontsize= 35)
    #fig.savefig('figures/fig3.pdf')
    fig.savefig('comparison_'+str(L)+'_'+str(SEED)+'.jpg', dpi=600)
```
